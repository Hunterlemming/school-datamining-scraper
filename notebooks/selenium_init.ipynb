### Selenium telepítése
#### 1. Hozzunk léptre egy python virtuális környezetet, a projetünk mappájában nyitott terminállal:
        python3 -m venv my_venv_name
#### 2. Lépjünk be a környezetbe:
        source my_venv_name/bin/activate
#### 3. Adjuk hozzá a selenium packaget:
        pip install selenium
#### 4. Töltsük le az általunk használt böngésző megfelelő driverét:
- Chrome: &nbsp;https://sites.google.com/a/chromium.org/chromedriver/downloads
- Edge: &nbsp;&nbsp;&nbsp;https://developer.microsoft.com/en-us/microsoft-edge/tools/webdriver/
- Firefox: https://github.com/mozilla/geckodriver/releases
- Safari: &nbsp;https://webkit.org/blog/6900/webdriver-support-in-safari-10/
#### 5. Helyezzük a driver-t egy olyan mappába amit a program bármikor el tud érni (pl.: a project mappájába)!

### Selenium import

```python
from selenium import webdriver                                  # Kötelező import - ez a Selenium maga
from selenium.webdriver.chrome.options import Options           # Opcionális import - a driver beállításai (én Chrome-ot használok)
from selenium.common.exceptions import NoSuchElementException   # Opcionális import - scrapeléskor ilyen hibát dob, ha nem talál valamit -> hibakezelés
```

### Selenium indítása

```python
options = Options()
options.headless = False    # Headless futtatási beállítás (debugoláskor érdemes kikapcsolni, hogy nyomon tudjuk követni az eseményeket - egyébként lassít)
DRIVER = webdriver.Chrome(options=options, executable_path="../selenium_driver/chromedriver")
```

```python
DRIVER.get("https://pypi.org/project/selenium/")
```

```python
DRIVER.close()
```
