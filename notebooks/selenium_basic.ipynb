# XPATH
Előbb-utóbb látni fogjuk, hogy ezek az attribute-alapú függvények eléggé limitáltak és komplexebb (vagy csak simán igénytelen) oldalakon már nem lesznek elegek.
## Mi is az XPATH?
Az "XML Path Language" rövidítése, mely egy olyan nyelvet takar ami egy "út szerű" szintaktikát használva navigál az oldal (vagy xml) elemei (node-ok) között.
### Az előző contentArea class-os dolog megvalósítása XPATH-el:

```python
cAces_by_xpath = DRIVER.find_elements_by_xpath("//div[@class='contentArea']")
print(f"Number of elements with class='contentArea': {len(cAces_by_xpath)}")
fcAce_by_xpath = cAces_by_xpath[0]
print(f"Text of first:\n{fcAce_by_xpath.text}")
```

## Felépítése:
        //div[@class='contentArea']
### "//", "/"
ROOT: Keresés kiinduló nodeja/elementje. Ha nincsen megadva semmi a html dokumentum-ot veszi annak.
- //: az útvonal relatív - tehát "Keress a ROOT-ból belül bárhol egy olyan divet ahol ..."
- /: &nbsp;az útvonal direkt - tehát "Keress a ROOT közvetlen gyerek-elementje között olyat, ahol ..."

### div (vagy bármely más html element)
A keresett element típusa. Ha láncoljuk az XPATH-t, az utolsó.

### \[@attribute='restriction'\]
- \[\]-el jelöljük, ha az előző element valamelyik tulajdonságá szerint szeretnénk szűkíteni a keresést.
- @attribute: A tulajdonság neve "@" jel után.
- 'restriction': ''-ök között a tulajdonság megkötése.

```python
offer_sections_in_content_area = DRIVER.find_elements_by_xpath("//div[@class='contentArea']/section[@itemprop='offers']")
print("Offers:\n")
for offer in offer_sections_in_content_area:
    print(offer.text)
```

### Itt egy utat írtunk le, ami:
&nbsp;&nbsp;olyan **<section\>**-öket fog keresni<br>&nbsp;&nbsp;az oldal **<div class="contentArea"\>** element-jeinek<br>&nbsp;&nbsp;&nbsp;&nbsp;**elsőfokú gyerek**-elementjei között,<br>&nbsp;&nbsp;melyeknek az **itemprop attribútumának értéke "offers"**.

```python
first_offer_section = offer_sections_in_content_area[0]
price_value_span = first_offer_section.find_element_by_xpath(".//span[contains(@class,'price')]")
currency_span = price_value_span.find_element_by_xpath("(./following-sibling::span)[1]")
```

### Itt az előző elementet vettük kiindulási pontnak (ROOT):
#### "/" és "//" helyett "./"-t és ".//"-t használunk!
Ezzel jelezzük, hogy nem az HTML-DOCUMENT, hanem a változóban tárolt WebElement a ROOT-ja a következő XPATH-nek.
#### Keresési függvény (szerűség?).
A \[\]-eken belül különböző XPATH függvényeket is meghívhatunk a saját szintaktikájukkal. A contains például egy szövegrészletet keres a megadott attribútum értékében.<br>
Ha az előtte álló XPATH-t \(\)-kel kijelöljük, listaként kezelve lekérhetjük n-edik adattagját. XPATH-ben 0, helyett 1 az első index!
#### element_A/following-sibling::elementType
A HTML hard-code-ban az element_A-t követő, az ő globális szintjén lévő (pl.: HTML-DOC ROOT-tól számítva minden harmad-fokú gyerek) elementType típusú objektumok listája.

```python
first_price = {}
first_price["value"] = int(price_value_span.text.replace(" ", ""))
first_price["currency"] = currency_span.text.replace(" ", "")
print(first_price)
```

### Oldalak között válthatunk úgy, hogy megkeressük a következő page-url-t tartalmazó objektumot és átadjuk a href attribútumát a DRIVER.get()-nek:

```python
next_page_link = DRIVER.find_element_by_xpath("//a[contains(@class,'jofogasicon-right')]")
next_url = next_page_link.get_attribute("href")
print(f"Going to: {next_url}")
DRIVER.get(next_url)
```

### Vagy egyszerűen kattintsunk a linket tartalmazó objektumra:

```python
next_page_link = DRIVER.find_element_by_xpath("//a[contains(@class,'jofogasicon-right')]")
print("Clicking next button")
DRIVER.execute_script("arguments[0].click();", next_page_link)
```
